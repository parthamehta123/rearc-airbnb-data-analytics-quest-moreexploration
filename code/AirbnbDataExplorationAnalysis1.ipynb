***5. Handling Missing Values***

***There are quite a few columns with NaN / Missing Values. I will adress them all individually.***

```python
# Checking how much data is missing

df_drop.isnull().sum()
```

```python
df_drop.dropna(subset = ['host_is_superhost'], inplace = True)
```

***Since there are no NaN in 'host_since', I am directly converting it from object to datetime***

```python
df_drop['host_since'] = pd.to_datetime(df_drop['host_since'])
```

***Replacing NaN from 'host_response_time' & 'bathrooms_text' to 'Unknown'***

Note: I decided to impute 'Unknown' as these missing rows belong to text-based, categorical columns

```python
df_drop['host_response_time'].fillna('Unknown', inplace=True)
df_drop['bathrooms_text'].fillna('Unknown', inplace=True) 
```

***Handling 'host_response_rate' & 'host_acceptance_rate' columns***

Note: Here is a slightly complex transformation, but necessary in order to be able to correlate these values

1. Removing the % sign from the 'host_response_rate' & 'host_acceptance_rate' columns

```python
df_drop['host_response_rate'] = df_drop['host_response_rate'].str.replace('%','')
```

```python
df_drop['host_acceptance_rate'] = df_drop['host_acceptance_rate'].str.replace('%','')
```

2. Renaming the columns

```python
df_drop.rename(columns = {'host_response_rate' : 'host_resp_rate_pct', 'host_acceptance_rate' : 'host_accept_rate_pct'}, inplace = True)
```

3. Imputing the mean value for both columns (For ease, the mean was calculated in excel / Data type error in Python)

```python
df_drop['host_resp_rate_pct'].fillna(88, inplace=True)
```

```python
df_drop['host_accept_rate_pct'].fillna(88, inplace=True)
```

4. Changing the data type (from object to int)

```python
df_drop['host_resp_rate_pct'] = df_drop['host_resp_rate_pct'].astype('int')
```

```python
df_drop['host_accept_rate_pct'] = df_drop['host_accept_rate_pct'].astype('int')
```

***Handling the 'price' column***

The price column is read as Object due to the $ sign. I would like it to be in float format.

1. Removing the $ sign from the 'price' column

```python
df_drop['price'] = df_drop['price'].str.replace('$','', regex=True)
```

2. Renaming the columns

```python
df_drop.rename(columns = {'price' : 'price_usd'}, inplace = True)
```

3. Removing the ',' to be able to convert data type

```python
df_drop['price_usd'] = df_drop['price_usd'].str.replace(',','')
```

4. Changing the data type (from object to int)

```python
df_drop['price_usd'] = df_drop['price_usd'].astype(float)
```

***Replacing NaN from 'first_review' & 'last_review' to '1900-01-01'***

Note: I decided to impute '1900-01-01' as it will be easy to see these dates as being just place holders and it will allow me to convert all of the columns values into a date format.

```python
df_drop['first_review'].fillna('1900-01-01', inplace=True)
df_drop['last_review'].fillna('1900-01-01', inplace=True)
```

Now converting the data type to date

```python
df_drop['first_review'] = pd.to_datetime(df_drop['first_review']) 
df_drop['last_review'] = pd.to_datetime(df_drop['last_review']) 
```

***Imputing the mean for 'bedrooms', 'beds' and other continuous variables***

Note: Here I made this choice to be able to use this dataset and find correlations, without changing key metrics.

For 'bedrooms' and 'beds', imputing the mean allows me to still know which values are imputed and which are orginal. For all other continuous variables, imputing the mean won't change the dataset's key metrics but will improve usability.

```python
print(len(df_drop[df_drop.review_scores_rating == 0]))
print(len(df_drop[df_drop.review_scores_accuracy == 0]))
print(len(df_drop[df_drop.review_scores_cleanliness == 0]))
print(len(df_drop[df_drop.review_scores_checkin == 0]))
print(len(df_drop[df_drop.review_scores_communication == 0]))
print(len(df_drop[df_drop.review_scores_location == 0]))
print(len(df_drop[df_drop.review_scores_value == 0]))
print(len(df_drop[df_drop.reviews_per_month == 0]))
print(len(df_drop[df_drop.accommodates == 0]))
print(len(df_drop[df_drop.price_usd == 0]))
print(len(df_drop[df_drop.minimum_nights > 365]))
```
